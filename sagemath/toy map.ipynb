```python
var("a, φ, x", domain='real')
```

```python
f(x)=sgn(x)*a*cos(abs(x)+φ)
```

```python
b = a*cos(φ)
```

```python
solve(abs(f(b))==0,φ)
```

```python
find_root(abs(f.subs(a=2.6)(b.subs(a=2.6)))==b.subs(a=2.6),0,pi)
```
